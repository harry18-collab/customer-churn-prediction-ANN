```python
import tensorflow as tf
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model
import pickle
```

```python
model=load_model('model.h5')

with open('Label_encoder_gender.pkl','rb') as file:
    label_encoder_gender=pickle.load(file)

with open('OneHotEncoder_geo.pkl','rb') as file:
    OneHotEncoder_geo=pickle.load(file)

with open('StandardScaler','rb') as file:
    scaler=pickle.load(file)
```

```python
input_data_1 = {
    'CreditScore': 400,
    'Geography': 'Germany',
    'Gender': 'Female',
    'Age': 55,
    'Tenure': 1,
    'Balance': 90000,
    'NumOfProducts': 1,
    'HasCrCard': 0,
    'IsActiveMember': 0,
    'EstimatedSalary': 30000
}
```

```python
input_df_1=pd.DataFrame([input_data_1])
input_df_1
```

```python
geo_encoded=OneHotEncoder_geo.transform([[input_data_1['Geography']]]).toarray()
geo_df=pd.DataFrame(geo_encoded,columns=OneHotEncoder_geo.get_feature_names_out(['Geography']))
geo_df
```

```python
#encode categorical value
input_df_1['Gender']=label_encoder_gender.transform([input_data_1['Gender']])
input_df_1
```

```python
#concatinate the df
input_df_1=pd.concat([input_df_1.drop('Geography',axis=1),geo_df],axis=1)
input_df_1
```

```python
#scaling
scaled_input_1=scaler.transform(input_df_1)
scaled_input_1
```

```python
prediction=model.predict(scaled_input_1)
prediction_prob=prediction[0][0]
print(prediction_prob)
```

```python
if(prediction_prob>0.5):
    print(f'the customer will exit the bank')
else :
    print(f'person will not exit the bank')
```
